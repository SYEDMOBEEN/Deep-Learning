```python
import numpy as np
import pandas as pd
from keras import models
from keras import layers
from sklearn.utils import shuffle
```

```python
sonar_data = pd.read_csv("sonar.csv" , header = None) 
```

```python
data = sonar_data
```

```python
data = shuffle(data)
```

```python
data.shape
```

```python
data.head()
```

```python
y_lables = data.iloc[:,-1]
```

```python
y_lables = y_lables.replace({"R":"1","M":"0"})
```

```python
y_lables
```

```python
x_data = data.iloc[:,0:60]
```

```python
x_data.shape
```

```python
x_data.head()
```

```python
x_train=x_data.iloc[:int(x_data.shape[0]*.70),:]
```

```python
x_train.shape
```

```python
x_test=x_data.iloc[int(x_data.shape[0]*.70):,:]
```

```python
x_test.shape
```

```python
y_train=y_lables[:145]
```

```python
y_test=y_lables[145:]
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(60,)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam',
loss='binary_crossentropy',
metrics=['accuracy'])
```

```python
x_train = x_train.astype('float32')
y_train = y_train.astype('float32')
```

```python
history = model.fit(x_train,y_train,epochs=100,batch_size=512)
```

```python
x_test = x_test.astype('float32')
y_test = y_test.astype('float32')
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print('test_acc:', test_acc)
```
